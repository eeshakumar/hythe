```python

br_iqn_baseline = pd.read_pickle(os.path.join(exp_dir_root, "exp_iqn_default/benchmark/data"))
br_iqn_2split = pd.read_pickle(os.path.join(exp_dir_root, "exp_2_lbnw_ld/benchmark/data"))
br_iqn_4split = pd.read_pickle(os.path.join(exp_dir_root, "exp_4_lbnw_ld/benchmark/data"))
br_iqn_2split_th = pd.read_pickle(os.path.join(exp_dir_root, "exp_2_ldth/benchmark/data"))
br_iqn_4split_th = pd.read_pickle(os.path.join(exp_dir_root, "exp_4_ldth/benchmark/data"))
br_iqn_2split_di64 = pd.read_pickle(os.path.join(exp_dir_root, "exp_2_lddi64/benchmark/data"))
br_iqn_4split_th = pd.read_pickle(os.path.join(exp_dir_root, "exp_4_lddi64/benchmark/data"))


br_iqn_baseline.columns

br_iqn_2split.head(100)
```

```python
plt.rcParams['figure.figsize'] = (40.0, 20.0)
plt.rcParams['font.family'] = "serif"

# concatenate to identify columns by dataset column
concatenated = pd.concat([br_iqn_baseline.assign(dataset='iqn_vanilla'), br_iqn_2split.assign(dataset='iqn_2split'),
                          br_iqn_4split.assign(dataset='iqn_4split'), br_iqn_4split.assign(dataset='iqn_2split_th'),
                          br_iqn_4split.assign(dataset='iqn_4split_th'), br_iqn_4split.assign(dataset='iqn_2split_di64'),
                          br_iqn_4split.assign(dataset='iqn_4split_di64')])

# sns.countplot(data=br_iqn_baseline, x = 'collision_other')
sns_plot = sns.countplot(data=concatenated, x = 'dataset', hue='collision_other')
sns_plot.set(xlabel='Dataset', ylabel = "Collisions")
plt.show()

sns_plot.figure.savefig("collision_other.png")

# similarly plot success

plt.figure()

sns_plot = sns.countplot(data=concatenated, x = 'dataset', hue='success')
sns_plot.set(xlabel='Dataset', ylabel = "Success")

plt.show()

sns_plot.figure.savefig("success.png")


# similarly plot max steps taken

plt.figure()

sns_plot = sns.countplot(data=concatenated, x = 'dataset', hue='max_steps')
sns_plot.set(xlabel='Dataset', ylabel = "Max Steps in Scenario")
plt.show()

sns_plot.figure.savefig("max_steps.png")
```
